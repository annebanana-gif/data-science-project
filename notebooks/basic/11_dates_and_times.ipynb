# Working with Dates and Times

Handling dates and times is essential in data science. Python's `datetime` module and pandas provide powerful tools for temporal data.

**Key modules:**
- `datetime` - Python's built-in date/time handling
- `pandas` - Advanced datetime functionality for data analysis

## Python's `datetime` Module

The datetime module provides classes for working with dates and times.

### Creating Date and Time Objects

```python
from datetime import datetime, date, time, timedelta

# Current date and time
now = datetime.now()
print('Current datetime:', now)

# Current date only
today = date.today()
print('Today:', today)

# Specific date
birthday = date(1990, 5, 15)
print('Birthday:', birthday)

# Specific datetime
meeting = datetime(2026, 2, 15, 14, 30, 0)
print('Meeting:', meeting)

# Time only
lunch_time = time(12, 30, 0)
print('Lunch time:', lunch_time)
```

### Accessing Date Components

```python
now = datetime.now()

print(f'Year: {now.year}')
print(f'Month: {now.month}')
print(f'Day: {now.day}')
print(f'Hour: {now.hour}')
print(f'Minute: {now.minute}')
print(f'Second: {now.second}')
print(f'Weekday: {now.weekday()}')  # Monday = 0, Sunday = 6
print(f'ISO Weekday: {now.isoweekday()}')  # Monday = 1, Sunday = 7
```

### Formatting Dates - `strftime()`

Convert datetime objects to strings with custom formats.

```python
now = datetime.now()

# Common formats
print('Default:', now)
print('ISO format:', now.isoformat())
print('Date only:', now.strftime('%Y-%m-%d'))
print('Time only:', now.strftime('%H:%M:%S'))
print('US format:', now.strftime('%m/%d/%Y'))
print('European format:', now.strftime('%d/%m/%Y'))
print('Full text:', now.strftime('%A, %B %d, %Y at %I:%M %p'))
print('Short text:', now.strftime('%a, %b %d, %Y'))

# Common format codes:
# %Y - 4-digit year    %m - Month (01-12)    %d - Day (01-31)
# %H - Hour (00-23)    %M - Minute (00-59)   %S - Second (00-59)
# %A - Full weekday    %a - Short weekday    %B - Full month
# %b - Short month     %I - Hour (01-12)     %p - AM/PM
```

### Parsing Dates - `strptime()`

Convert strings to datetime objects.

```python
# Parse different date formats
date_str1 = '2026-02-15'
dt1 = datetime.strptime(date_str1, '%Y-%m-%d')
print(f'Parsed: {dt1}')

date_str2 = '15/02/2026 14:30'
dt2 = datetime.strptime(date_str2, '%d/%m/%Y %H:%M')
print(f'Parsed: {dt2}')

date_str3 = 'February 15, 2026'
dt3 = datetime.strptime(date_str3, '%B %d, %Y')
print(f'Parsed: {dt3}')

# Handle parsing errors
try:
    bad_date = datetime.strptime('invalid', '%Y-%m-%d')
except ValueError as e:
    print(f'Parse error: {e}')
```

## Time Arithmetic with `timedelta`

Perform calculations with dates and times.

```python
from datetime import timedelta

now = datetime.now()

# Add/subtract time
tomorrow = now + timedelta(days=1)
yesterday = now - timedelta(days=1)
next_week = now + timedelta(weeks=1)
in_2_hours = now + timedelta(hours=2)

print(f'Now: {now}')
print(f'Tomorrow: {tomorrow}')
print(f'Yesterday: {yesterday}')
print(f'Next week: {next_week}')
print(f'In 2 hours: {in_2_hours}')

# Complex timedelta
future = now + timedelta(days=7, hours=3, minutes=30)
print(f'7 days, 3.5 hours from now: {future}')
```

```python
# Calculate time differences
event_date = datetime(2026, 12, 25, 0, 0, 0)
now = datetime.now()
time_until = event_date - now

print(f'Time until Christmas: {time_until}')
print(f'Days: {time_until.days}')
print(f'Total seconds: {time_until.total_seconds()}')
print(f'Hours: {time_until.total_seconds() / 3600:.1f}')
```

## Comparing Dates

```python
date1 = datetime(2026, 1, 1)
date2 = datetime(2026, 12, 31)
date3 = datetime(2026, 1, 1)

print(f'date1 < date2: {date1 < date2}')
print(f'date1 > date2: {date1 > date2}')
print(f'date1 == date3: {date1 == date3}')
print(f'date1 != date2: {date1 != date2}')

# Find min/max
dates = [
    datetime(2026, 3, 15),
    datetime(2026, 1, 10),
    datetime(2026, 6, 20)
]
print(f'Earliest: {min(dates)}')
print(f'Latest: {max(dates)}')
```

## Pandas DateTime Functionality

Pandas provides powerful tools for working with datetime data in DataFrames.

```python
import pandas as pd

# Create DataFrame with dates
dates = pd.date_range('2026-01-01', periods=10, freq='D')
df = pd.DataFrame({
    'date': dates,
    'value': range(10, 20)
})
print(df.head())
```

### Converting to Datetime

```python
# String to datetime
df_str = pd.DataFrame({
    'date_str': ['2026-01-01', '2026-01-02', '2026-01-03'],
    'value': [10, 20, 30]
})

# Convert to datetime
df_str['date'] = pd.to_datetime(df_str['date_str'])
print(df_str)
print(f'\nDate column type: {df_str["date"].dtype}')
```

```python
# Handle different formats
dates_mixed = pd.Series([
    '2026-01-01',
    '01/02/2026',
    'Feb 15, 2026'
])

# Try to infer format
parsed = pd.to_datetime(dates_mixed, infer_datetime_format=True, errors='coerce')
print(parsed)
```

### Extracting Date Components

```python
# Create sample data
dates = pd.date_range('2026-01-01', periods=100, freq='D')
df = pd.DataFrame({'date': dates, 'value': range(100)})

# Extract components
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.day_name()
df['week'] = df['date'].dt.isocalendar().week
df['quarter'] = df['date'].dt.quarter

print(df.head(10))
```

### Date Ranges and Frequencies

```python
# Daily frequency
daily = pd.date_range('2026-01-01', periods=7, freq='D')
print('Daily:', daily)

# Business days (Mon-Fri)
business = pd.date_range('2026-01-01', periods=7, freq='B')
print('\nBusiness days:', business)

# Weekly on Mondays
weekly = pd.date_range('2026-01-01', periods=4, freq='W-MON')
print('\nWeekly:', weekly)

# Monthly - end of month
monthly = pd.date_range('2026-01-01', periods=6, freq='M')
print('\nMonthly:', monthly)

# Hourly
hourly = pd.date_range('2026-01-01', periods=5, freq='H')
print('\nHourly:', hourly)

# Common frequencies: D (day), W (week), M (month), Y (year),
# H (hour), T (minute), S (second), B (business day)
```

### Filtering by Date

```python
# Create sample data
dates = pd.date_range('2026-01-01', periods=365, freq='D')
df = pd.DataFrame({
    'date': dates,
    'value': range(365)
})

# Filter for specific month
january = df[df['date'].dt.month == 1]
print(f'January data: {len(january)} rows')

# Filter for date range
q1 = df[(df['date'] >= '2026-01-01') & (df['date'] <= '2026-03-31')]
print(f'Q1 data: {len(q1)} rows')

# Filter for weekends
weekends = df[df['date'].dt.dayofweek >= 5]
print(f'Weekend data: {len(weekends)} rows')
```

### Resampling Time Series

```python
# Create daily data
dates = pd.date_range('2026-01-01', periods=90, freq='D')
df = pd.DataFrame({
    'date': dates,
    'sales': range(100, 190)
})
df.set_index('date', inplace=True)

# Resample to weekly (sum)
weekly = df.resample('W').sum()
print('Weekly totals:')
print(weekly.head())

# Resample to monthly (mean)
monthly = df.resample('M').mean()
print('\nMonthly averages:')
print(monthly)
```

### Time Zones

```python
# Create timezone-aware datetime
utc_time = pd.Timestamp('2026-02-15 12:00:00', tz='UTC')
print(f'UTC: {utc_time}')

# Convert to different timezone
copenhagen = utc_time.tz_convert('Europe/Copenhagen')
print(f'Copenhagen: {copenhagen}')

new_york = utc_time.tz_convert('America/New_York')
print(f'New York: {new_york}')

tokyo = utc_time.tz_convert('Asia/Tokyo')
print(f'Tokyo: {tokyo}')
```

## Practical Examples

### Example 1: Calculate Age

```python
def calculate_age(birth_date):
    """Calculate age in years from birth date."""
    today = date.today()
    
    # Convert string to date if needed
    if isinstance(birth_date, str):
        birth_date = datetime.strptime(birth_date, '%Y-%m-%d').date()
    
    age = today.year - birth_date.year
    
    # Adjust if birthday hasn't occurred this year
    if today.month < birth_date.month or \
       (today.month == birth_date.month and today.day < birth_date.day):
        age -= 1
    
    return age

print(f'Age: {calculate_age("1990-05-15")} years')
print(f'Age: {calculate_age("2000-12-25")} years')
```

### Example 2: Business Days Between Dates

```python
def business_days_between(start, end):
    """Count business days between two dates."""
    start = pd.to_datetime(start)
    end = pd.to_datetime(end)
    
    # Generate business day range
    business_days = pd.bdate_range(start, end)
    
    return len(business_days)

start = '2026-01-01'
end = '2026-01-31'
days = business_days_between(start, end)
print(f'Business days from {start} to {end}: {days}')
```

| `%a` | Short weekday | Mon |

| `%M` | Minute (00-59) | 30 |

| `%p` | AM/PM | PM |

## Best Practices

1. **Always use timezone-aware datetimes** for production systems
2. **Store in UTC**, display in local timezone
3. **Use ISO format** (YYYY-MM-DD) for consistency
4. **Validate date inputs** before parsing
5. **Handle missing/invalid dates** gracefully with try-except
6. **Use pandas for large datasets** - it's optimized for performance
7. **Consider business days** vs calendar days when appropriate
8. **Document date formats** clearly in your code
