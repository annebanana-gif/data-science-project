# Object-Oriented Programming (OOP)

Object-Oriented Programming is a programming paradigm that organizes code around **objects** and **classes**. It helps create reusable, modular, and maintainable code.

**Key Concepts:**
- **Class**: A blueprint for creating objects (like a recipe)
- **Object**: An instance of a class (like a cake made from the recipe)
- **Attributes**: Data stored in objects (variables)
- **Methods**: Functions that belong to objects

**Why OOP?**
- Code reusability
- Better organization
- Easier to maintain and extend
- Models real-world entities naturally

## Creating Your First Class

````python
# Simple class definition
class Dog:
    pass

# Create an object (instance)
my_dog = Dog()
print(type(my_dog))
print(my_dog)
````

## Adding Attributes

Attributes are variables that belong to an object.

````python
class Dog:
    # Class attribute (shared by all instances)
    species = "Canis familiaris"
    
    # Constructor method (runs when object is created)
    def __init__(self, name, age):
        # Instance attributes (unique to each instance)
        self.name = name
        self.age = age

# Create dog objects
buddy = Dog("Buddy", 5)
luna = Dog("Luna", 3)

print(f"{buddy.name} is {buddy.age} years old")
print(f"{luna.name} is {luna.age} years old")
print(f"Both are {buddy.species}")
````

## Understanding `self`

- `self` refers to the instance of the class
- It's automatically passed as the first parameter to instance methods
- You can name it anything, but `self` is the convention

````python
class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        # self.name refers to this specific person's name
        return f"Hello, I'm {self.name}"

alice = Person("Alice")
bob = Person("Bob")

print(alice.greet())  # self = alice
print(bob.greet())    # self = bob
````

## Adding Methods

Methods are functions that belong to a class.

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        return f"{self.name} says Woof!"
    
    def get_human_age(self):
        return self.age * 7
    
    def have_birthday(self):
        self.age += 1
        return f"Happy birthday {self.name}! Now {self.age} years old."

buddy = Dog("Buddy", 5)
print(buddy.bark())
print(f"In human years: {buddy.get_human_age()}")
print(buddy.have_birthday())
print(f"Age now: {buddy.age}")
````

## Class vs Instance Variables

````python
class BankAccount:
    # Class variable (shared by all accounts)
    interest_rate = 0.02
    num_accounts = 0
    
    def __init__(self, owner, balance=0):
        # Instance variables (unique to each account)
        self.owner = owner
        self.balance = balance
        BankAccount.num_accounts += 1
    
    def apply_interest(self):
        self.balance *= (1 + BankAccount.interest_rate)

# Create accounts
acc1 = BankAccount("Alice", 1000)
acc2 = BankAccount("Bob", 2000)

print(f"Total accounts: {BankAccount.num_accounts}")
print(f"Alice's balance: ${acc1.balance}")
print(f"Bob's balance: ${acc2.balance}")

# Change class variable
BankAccount.interest_rate = 0.03
acc1.apply_interest()
acc2.apply_interest()

print(f"\nAfter interest:")
print(f"Alice's balance: ${acc1.balance:.2f}")
print(f"Bob's balance: ${acc2.balance:.2f}")
````

## Special Methods (Dunder Methods)

Special methods have double underscores before and after the name (`__method__`). They enable special Python features.

### `__str__` and `__repr__`

````python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages
    
    def __str__(self):
        # User-friendly string representation
        return f"'{self.title}' by {self.author}"
    
    def __repr__(self):
        # Developer-friendly representation (should be unambiguous)
        return f"Book('{self.title}', '{self.author}', {self.pages})"

book = Book("1984", "George Orwell", 328)

print(str(book))   # Calls __str__
print(repr(book))  # Calls __repr__
print(book)        # Uses __str__ if available
````

### Other Useful Special Methods

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return f"Vector({self.x}, {self.y})"
    
    # Addition
    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)
    
    # Length (for len())
    def __len__(self):
        return 2
    
    # Equality
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
    # Absolute value
    def __abs__(self):
        return (self.x**2 + self.y**2)**0.5

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1: {v1}")
print(f"v2: {v2}")
print(f"v1 + v2: {v1 + v2}")
print(f"Length of v1: {len(v1)}")
print(f"Magnitude of v1: {abs(v1)}")
print(f"v1 == v2: {v1 == v2}")
````

## Inheritance

Inheritance allows a class to inherit attributes and methods from another class.

- **Parent/Base/Super class**: The class being inherited from
- **Child/Derived/Sub class**: The class that inherits

````python
# Parent class
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def make_sound(self):
        return "Some generic sound"
    
    def info(self):
        return f"{self.name} is a {self.species}"

# Child class
class Dog(Animal):
    def __init__(self, name, breed):
        # Call parent's __init__
        super().__init__(name, species="Dog")
        self.breed = breed
    
    # Override parent method
    def make_sound(self):
        return "Woof!"
    
    # Add new method
    def fetch(self):
        return f"{self.name} is fetching the ball!"

class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, species="Cat")
        self.color = color
    
    def make_sound(self):
        return "Meow!"

# Create instances
dog = Dog("Buddy", "Golden Retriever")
cat = Cat("Whiskers", "Orange")

print(dog.info())
print(dog.make_sound())
print(dog.fetch())

print(f"\n{cat.info()}")
print(cat.make_sound())
print(f"{cat.name} is {cat.color}")
````

## Encapsulation

Encapsulation is about hiding internal details and providing controlled access to data.

**Convention in Python:**
- `public_var`: Public (accessible from anywhere)
- `_protected_var`: Protected (convention: use only within class and subclasses)
- `__private_var`: Private (name mangling makes it harder to access)

````python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner           # Public
        self._account_id = 12345     # Protected
        self.__balance = balance     # Private
    
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            return f"Deposited ${amount}. New balance: ${self.__balance}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            return f"Withdrew ${amount}. New balance: ${self.__balance}"
        return "Insufficient funds or invalid amount"
    
    def get_balance(self):
        return self.__balance

account = BankAccount("Alice", 1000)

print(account.owner)           # ✅ OK: public
print(account._account_id)     # ⚠️ Works but discouraged
# print(account.__balance)     # ❌ Error: AttributeError

print(account.get_balance())   # ✅ Proper way to access private data
print(account.deposit(500))
print(account.withdraw(200))
````

## Property Decorators

Properties allow you to use getter and setter methods with attribute-like syntax.

````python
class Temperature:
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        """Get temperature in Celsius"""
        return self._celsius
    
    @celsius.setter
    def celsius(self, value):
        """Set temperature in Celsius"""
        if value < -273.15:
            raise ValueError("Temperature below absolute zero!")
        self._celsius = value
    
    @property
    def fahrenheit(self):
        """Get temperature in Fahrenheit"""
        return (self._celsius * 9/5) + 32
    
    @fahrenheit.setter
    def fahrenheit(self, value):
        """Set temperature using Fahrenheit"""
        self.celsius = (value - 32) * 5/9

# Create temperature object
temp = Temperature(25)

# Use like attributes, but getters/setters run behind the scenes
print(f"Temperature: {temp.celsius}°C")
print(f"Temperature: {temp.fahrenheit}°F")

temp.fahrenheit = 100
print(f"\nAfter setting to 100°F:")
print(f"Celsius: {temp.celsius:.1f}°C")
print(f"Fahrenheit: {temp.fahrenheit}°F")

# Validation works
try:
    temp.celsius = -300
except ValueError as e:
    print(f"\nError: {e}")
````

## Class Methods and Static Methods

````python
class Date:
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
    
    def __str__(self):
        return f"{self.year}-{self.month:02d}-{self.day:02d}"
    
    @classmethod
    def from_string(cls, date_string):
        """Alternative constructor (class method)"""
        year, month, day = map(int, date_string.split('-'))
        return cls(year, month, day)
    
    @classmethod
    def today(cls):
        """Create date for today"""
        import datetime
        today = datetime.date.today()
        return cls(today.year, today.month, today.day)
    
    @staticmethod
    def is_leap_year(year):
        """Static method (doesn't need instance or class)"""
        return year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)

# Regular constructor
date1 = Date(2026, 2, 11)
print(f"date1: {date1}")

# Class method (alternative constructor)
date2 = Date.from_string("2025-12-25")
print(f"date2: {date2}")

# Class method
date3 = Date.today()
print(f"Today: {date3}")

# Static method
print(f"Is 2024 a leap year? {Date.is_leap_year(2024)}")
print(f"Is 2026 a leap year? {Date.is_leap_year(2026)}")
````

## Polymorphism

Polymorphism means "many forms" - different classes can implement the same method in different ways.

````python
class Shape:
    def area(self):
        raise NotImplementedError("Subclasses must implement area()")
    
    def perimeter(self):
        raise NotImplementedError("Subclasses must implement perimeter()")

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        import math
        return 2 * math.pi * self.radius

# Polymorphism in action
shapes = [
    Rectangle(5, 10),
    Circle(7),
    Rectangle(3, 3)
]

for i, shape in enumerate(shapes, 1):
    print(f"Shape {i}: Area = {shape.area():.2f}, Perimeter = {shape.perimeter():.2f}")
````

## Multiple Inheritance

A class can inherit from multiple parent classes.

````python
class Flyer:
    def fly(self):
        return "I can fly!"

class Swimmer:
    def swim(self):
        return "I can swim!"

class Duck(Flyer, Swimmer):
    def __init__(self, name):
        self.name = name
    
    def quack(self):
        return f"{self.name} says Quack!"

duck = Duck("Donald")
print(duck.quack())
print(duck.fly())
print(duck.swim())

# Method Resolution Order (MRO)
print(f"\nMRO: {[cls.__name__ for cls in Duck.__mro__]}")
````

## Practical Example: Data Science Class

````python
import numpy as np

class Dataset:
    """A class to manage and analyze datasets"""
    
    def __init__(self, name, data):
        self.name = name
        self.data = np.array(data)
        self._processed = False
    
    def __len__(self):
        return len(self.data)
    
    def __str__(self):
        return f"Dataset '{self.name}' with {len(self)} samples"
    
    def __repr__(self):
        return f"Dataset(name='{self.name}', samples={len(self)})"
    
    @property
    def mean(self):
        return self.data.mean()
    
    @property
    def std(self):
        return self.data.std()
    
    @property
    def median(self):
        return np.median(self.data)
    
    def normalize(self):
        """Z-score normalization"""
        self.data = (self.data - self.mean) / self.std
        self._processed = True
        return self
    
    def summary(self):
        """Get statistical summary"""
        return {
            'name': self.name,
            'count': len(self),
            'mean': self.mean,
            'std': self.std,
            'median': self.median,
            'min': self.data.min(),
            'max': self.data.max(),
            'processed': self._processed
        }
    
    @classmethod
    def from_random(cls, name, size, seed=None):
        """Create dataset with random data"""
        if seed:
            np.random.seed(seed)
        data = np.random.randn(size) * 10 + 50
        return cls(name, data)

# Create and use the dataset
dataset = Dataset.from_random("Experiment_A", 100, seed=42)
print(dataset)
print(f"Length: {len(dataset)}")

print("\nBefore normalization:")
summary = dataset.summary()
for key, value in summary.items():
    if isinstance(value, float):
        print(f"{key}: {value:.2f}")
    else:
        print(f"{key}: {value}")

dataset.normalize()

print("\nAfter normalization:")
summary = dataset.summary()
for key, value in summary.items():
    if isinstance(value, float):
        print(f"{key}: {value:.2f}")
    else:
        print(f"{key}: {value}")
````

## Practical Example: Machine Learning Model Class

````python
class SimpleLinearRegression:
    """Simple Linear Regression: y = mx + b"""
    
    def __init__(self):
        self.slope = None
        self.intercept = None
        self._is_fitted = False
    
    def fit(self, X, y):
        """Train the model"""
        X = np.array(X)
        y = np.array(y)
        
        # Calculate slope and intercept
        x_mean = X.mean()
        y_mean = y.mean()
        
        numerator = np.sum((X - x_mean) * (y - y_mean))
        denominator = np.sum((X - x_mean) ** 2)
        
        self.slope = numerator / denominator
        self.intercept = y_mean - self.slope * x_mean
        self._is_fitted = True
        
        return self
    
    def predict(self, X):
        """Make predictions"""
        if not self._is_fitted:
            raise ValueError("Model not fitted yet. Call fit() first.")
        
        X = np.array(X)
        return self.slope * X + self.intercept
    
    def score(self, X, y):
        """Calculate R² score"""
        if not self._is_fitted:
            raise ValueError("Model not fitted yet. Call fit() first.")
        
        y = np.array(y)
        predictions = self.predict(X)
        
        ss_res = np.sum((y - predictions) ** 2)
        ss_tot = np.sum((y - y.mean()) ** 2)
        
        return 1 - (ss_res / ss_tot)
    
    def __str__(self):
        if self._is_fitted:
            return f"LinearRegression(y = {self.slope:.2f}x + {self.intercept:.2f})"
        return "LinearRegression(not fitted)"

# Example usage
np.random.seed(42)
X = np.linspace(0, 10, 50)
y = 2.5 * X + 5 + np.random.randn(50) * 2  # y = 2.5x + 5 + noise

# Create and train model
model = SimpleLinearRegression()
model.fit(X, y)

print(model)
print(f"R² Score: {model.score(X, y):.3f}")

# Make predictions
new_X = np.array([15, 20, 25])
predictions = model.predict(new_X)

print("\nPredictions:")
for x, pred in zip(new_X, predictions):
    print(f"  X = {x} → y = {pred:.2f}")
````

## Composition Over Inheritance

Sometimes it's better to use composition (has-a relationship) instead of inheritance (is-a relationship).

````python
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower
    
    def start(self):
        return "Engine started"
    
    def stop(self):
        return "Engine stopped"

class GPS:
    def __init__(self):
        self.location = (0, 0)
    
    def navigate(self, destination):
        return f"Navigating to {destination}"

class Car:
    """Car HAS-A engine and HAS-A GPS (composition)"""
    def __init__(self, model, horsepower):
        self.model = model
        self.engine = Engine(horsepower)  # Composition
        self.gps = GPS()                   # Composition
    
    def start(self):
        return f"{self.model}: {self.engine.start()}"
    
    def navigate_to(self, destination):
        return self.gps.navigate(destination)

car = Car("Tesla Model 3", 283)
print(car.start())
print(car.navigate_to("Aarhus University"))
print(f"Horsepower: {car.engine.horsepower}")
````

## Abstract Base Classes

Abstract classes cannot be instantiated and force subclasses to implement certain methods.

````python
from abc import ABC, abstractmethod

class DataProcessor(ABC):
    """Abstract base class for data processors"""
    
    @abstractmethod
    def process(self, data):
        """All subclasses must implement this method"""
        pass
    
    @abstractmethod
    def validate(self, data):
        """All subclasses must implement this method"""
        pass
    
    def log(self, message):
        """Concrete method (can be used as-is by subclasses)"""
        print(f"[LOG] {message}")

class CSVProcessor(DataProcessor):
    def process(self, data):
        self.log("Processing CSV data")
        return data.strip().split(',')
    
    def validate(self, data):
        return isinstance(data, str) and ',' in data

class JSONProcessor(DataProcessor):
    def process(self, data):
        self.log("Processing JSON data")
        import json
        return json.loads(data)
    
    def validate(self, data):
        import json
        try:
            json.loads(data)
            return True
        except:
            return False

# Cannot instantiate abstract class
# processor = DataProcessor()  # ❌ TypeError

# Can instantiate concrete classes
csv_proc = CSVProcessor()
json_proc = JSONProcessor()

csv_data = "Alice,30,Engineer"
print(f"Valid CSV? {csv_proc.validate(csv_data)}")
print(f"Processed: {csv_proc.process(csv_data)}")

json_data = '{"name": "Bob", "age": 25}'
print(f"\nValid JSON? {json_proc.validate(json_data)}")
print(f"Processed: {json_proc.process(json_data)}")
````

## Best Practices

1. **Follow naming conventions**
   - Class names: `PascalCase` (e.g., `BankAccount`, `DataProcessor`)
   - Methods/attributes: `snake_case` (e.g., `get_balance`, `user_name`)
   - Constants: `UPPER_CASE` (e.g., `MAX_SIZE`)

2. **Keep classes focused (Single Responsibility Principle)**
   ```python
   # ✅ GOOD: Each class has one job
   class EmailValidator:
       def validate(self, email): ...
   
   class EmailSender:
       def send(self, email, message): ...
   ```

3. **Use properties for computed attributes**
   ```python
   class Rectangle:
       def __init__(self, width, height):
           self.width = width
           self.height = height
       
       @property
       def area(self):
           return self.width * self.height
   ```

4. **Implement `__str__` and `__repr__`**
   ```python
   def __str__(self):
       return "User-friendly description"
   
   def __repr__(self):
       return "Developer-friendly representation"
   ```

5. **Use composition over inheritance when appropriate**
   - Prefer "has-a" over "is-a" relationships when it makes sense
   - Makes code more flexible and easier to test

6. **Document your classes**
   ```python
   class DataAnalyzer:
       """
       Analyzes numerical data.
       
       Parameters:
       -----------
       data : array-like
           The data to analyze
       
       Attributes:
       -----------
       mean : float
           The mean of the data
       """
   ```

7. **Don't overuse private variables**
   - Use `_protected` for internal use
   - Reserve `__private` for actual name conflicts
   - Python culture prefers "we're all consenting adults"
