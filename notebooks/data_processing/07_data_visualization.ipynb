# Data Visualization Basics

Data visualization is essential for exploring, understanding, and communicating insights from data. Python offers powerful libraries for creating visualizations.

**Key libraries:**
- **Matplotlib** - The foundation of Python visualization
- **Seaborn** - Statistical visualizations built on matplotlib
- **Pandas** - Built-in plotting capabilities

Install if needed: `poetry add matplotlib seaborn`

## Getting Started with Matplotlib

````python
import matplotlib.pyplot as plt
import numpy as np

# Basic line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True, alpha=0.3)
plt.show()
````

## Basic Plot Types

### Line Plot

````python
# Multiple lines
x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 6))
plt.plot(x, np.sin(x), label='sin(x)', linewidth=2)
plt.plot(x, np.cos(x), label='cos(x)', linewidth=2, linestyle='--')
plt.plot(x, np.sin(x) * np.cos(x), label='sin(x)*cos(x)', linewidth=2, linestyle=':')

plt.title('Trigonometric Functions', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()
````

### Scatter Plot

````python
# Random data
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
plt.colorbar(scatter, label='Color value')
plt.title('Scatter Plot with Varying Sizes and Colors', fontsize=14)
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()
````

### Bar Chart

````python
# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color='skyblue', edgecolor='navy', alpha=0.7)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

plt.title('Bar Chart Example', fontsize=14)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

### Histogram

````python
# Random data from normal distribution
data = np.random.randn(1000)

plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='coral', edgecolor='black', alpha=0.7)
plt.title('Histogram of Normal Distribution', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.axvline(data.mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {data.mean():.2f}')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
````

### Box Plot

````python
# Multiple datasets
data1 = np.random.normal(100, 10, 200)
data2 = np.random.normal(90, 20, 200)
data3 = np.random.normal(80, 15, 200)

data = [data1, data2, data3]
labels = ['Group A', 'Group B', 'Group C']

plt.figure(figsize=(10, 6))
box = plt.boxplot(data, labels=labels, patch_artist=True)

# Customize colors
colors = ['lightblue', 'lightgreen', 'lightcoral']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Box Plot Comparison', fontsize=14)
plt.ylabel('Values', fontsize=12)
plt.grid(axis='y', alpha=0.3)
plt.show()
````

````python
# Sample data
labels = ['Python', 'JavaScript', 'Java', 'C++', 'Others']
sizes = [35, 25, 15, 10, 15]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#ff99cc']
explode = (0.1, 0, 0, 0, 0)  # Explode first slice

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=90, explode=explode, shadow=True)
plt.title('Programming Language Distribution', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures circular pie
plt.show()
````

## Subplots - Multiple Plots in One Figure

````python
# Create 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Line
x = np.linspace(0, 10, 100)
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title('Line Plot')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2: Scatter
axes[0, 1].scatter(np.random.randn(50), np.random.randn(50), alpha=0.6)
axes[0, 1].set_title('Scatter Plot')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3: Bar
axes[1, 0].bar(['A', 'B', 'C'], [10, 20, 15])
axes[1, 0].set_title('Bar Chart')
axes[1, 0].grid(axis='y', alpha=0.3)

# Plot 4: Histogram
axes[1, 1].hist(np.random.randn(1000), bins=20, color='green', alpha=0.7)
axes[1, 1].set_title('Histogram')
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

## Seaborn - Statistical Visualizations

````python
import seaborn as sns
import pandas as pd

# Set style
sns.set_style('whitegrid')

# Load sample dataset
tips = sns.load_dataset('tips')
print(tips.head())
````

### Distribution Plot

````python
plt.figure(figsize=(12, 5))

# Histogram with KDE
plt.subplot(1, 2, 1)
sns.histplot(data=tips, x='total_bill', kde=True, color='skyblue')
plt.title('Distribution of Total Bill')

# KDE only
plt.subplot(1, 2, 2)
sns.kdeplot(data=tips, x='total_bill', fill=True, color='coral')
plt.title('Density Plot of Total Bill')

plt.tight_layout()
plt.show()
````

### Box Plot and Violin Plot

````python
plt.figure(figsize=(14, 5))

# Box plot
plt.subplot(1, 2, 1)
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
plt.title('Total Bill by Day and Gender')

# Violin plot (shows distribution)
plt.subplot(1, 2, 2)
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
plt.title('Total Bill Distribution by Day and Gender')

plt.tight_layout()
plt.show()
````

### Scatter Plot with Regression

````python
plt.figure(figsize=(10, 6))
sns.regplot(data=tips, x='total_bill', y='tip', scatter_kws={'alpha':0.5})
plt.title('Relationship between Total Bill and Tip', fontsize=14)
plt.show()
````

### Heatmap - Correlation Matrix

````python
# Calculate correlation matrix
numeric_cols = tips.select_dtypes(include=[np.number])
correlation = numeric_cols.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Correlation Matrix', fontsize=14)
plt.show()
````

### Pair Plot

````python
# Pairwise relationships
sns.pairplot(tips, hue='sex', diag_kind='kde', height=2.5)
plt.suptitle('Pairwise Relationships in Tips Dataset', y=1.02)
plt.show()
````

## Pandas Built-in Plotting

````python
import pandas as pd

# Create sample data
df = pd.DataFrame({
    'Month': pd.date_range('2026-01', periods=12, freq='M'),
    'Sales': [100, 120, 140, 130, 150, 170, 160, 180, 200, 190, 210, 220],
    'Expenses': [80, 85, 90, 88, 95, 100, 98, 105, 110, 108, 115, 120]
})

df.set_index('Month', inplace=True)

# Line plot
df.plot(figsize=(12, 6), marker='o', linewidth=2)
plt.title('Monthly Sales and Expenses', fontsize=14)
plt.ylabel('Amount ($)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(['Sales', 'Expenses'])
plt.show()
````

````python
# Bar plot
df.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Monthly Comparison', fontsize=14)
plt.ylabel('Amount ($)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(['Sales', 'Expenses'])
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

## Customization and Styling

````python
# Available styles
print('Available styles:', plt.style.available)

# Use a specific style
with plt.style.context('seaborn-v0_8-darkgrid'):
    plt.figure(figsize=(10, 6))
    x = np.linspace(0, 10, 100)
    plt.plot(x, np.sin(x), linewidth=3, label='sin(x)')
    plt.plot(x, np.cos(x), linewidth=3, label='cos(x)')
    plt.title('Plot with Dark Grid Style', fontsize=16)
    plt.xlabel('X', fontsize=12)
    plt.ylabel('Y', fontsize=12)
    plt.legend(fontsize=10)
    plt.show()
````

### Color Maps

````python
# Demonstrate different colormaps
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
cmaps = ['viridis', 'plasma', 'coolwarm', 'RdYlGn']

data = np.random.rand(10, 10)

for ax, cmap in zip(axes.flat, cmaps):
    im = ax.imshow(data, cmap=cmap)
    ax.set_title(f'Colormap: {cmap}')
    plt.colorbar(im, ax=ax)

plt.tight_layout()
plt.show()
````

## Best Practices

1. **Choose the right chart type** for your data
   - Line: trends over time
   - Bar: comparing categories
   - Scatter: relationships between variables
   - Histogram: distribution
   - Box: statistical summary

2. **Use clear labels and titles**
   - Always label axes
   - Add descriptive titles
   - Include units in labels

3. **Color wisely**
   - Use colorblind-friendly palettes
   - Limit colors (usually 3-7)
   - Ensure sufficient contrast

4. **Keep it simple**
   - Remove unnecessary elements
   - Use appropriate scales
   - Avoid 3D charts unless necessary

5. **Make it accessible**
   - Use large enough fonts
   - Add legends when needed
   - Consider printing in grayscale

6. **Adjust figure size** appropriately
   - `figsize=(width, height)` in inches
   - Standard: (10, 6) for most plots
   - Adjust based on content

7. **Use `tight_layout()`** to prevent overlap

8. **Save high-resolution** for publications
   - `dpi=300` minimum
   - `bbox_inches='tight'` to avoid clipping
